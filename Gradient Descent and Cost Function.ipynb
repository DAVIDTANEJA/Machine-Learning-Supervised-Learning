## -------------------------------------------

```python
import numpy as np

def gradient_descent(x,y):             # we have to find minima of slope
    m_curr = b_curr = 0                # here we starting from 0 taking both of mse and b current value=0
    iterations=1000                    # change also iterations 10000, and checks how much cost reduced
    n=len(x)                           # n- is length of data points    and taking x,y length of values are same
    learning_rate = 0.001              # change it acc. 0.07 , 0.09
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr                           # y=mx+c , Its mean squared error
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])       # we reduce cost , Its cost function (divided mse by 1/n)
        md = -(2/n)*sum(x*(y-y_predicted))                          # m-derivative , then we take derivative of above eqn.
        bd = -(2/n)*sum(y-y_predicted)                              # b-derivative , again derivative
        
        m_curr = m_curr - learning_rate * md        # step eqn. , new m derived
        b_curr = b_curr - learning_rate * bd        # same        
        
        print('m  {}, b {}, cost {},  iteration {}'.format(m_curr,b_curr,cost,i))
        
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)

# See in every iteration , cost is reducing it means we are reaching towards global minima
# When we have to stop, compare the previous cost to the current cost, if it is reducing then goto that otherwise stop there 
```
